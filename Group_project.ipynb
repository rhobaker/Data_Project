## To access a CSV file from the GitHub repository:

- Go to repository and click on the selected dataset.
- On the top-right of the dataset view, click the 'raw' button.
-This will display the raw contents of the CSV file. Copy the URL from the address bar of your browser.
- In Colab, create a string variable called url and assign the copied raw GitHub link to it.
- Use pandas read csv to store the data to a dataframe.


```python
url = 'https://raw.githubusercontent.com/rhobaker/Data_Project/main/ONS_adult_smoking_habits_in_great_britain/Table_1a-Table%201.csv'
```

```python
import pandas as pd

table1a = pd.read_csv(url)
```
