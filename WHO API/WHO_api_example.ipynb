DATASET: https://www.who.int/data/gho/indicator-metadata-registry/imr-details/prevalence-of-current-tobacco-use-(crude-adjusted-rate)

```python
# Get API from the WHO
url = "https://ghoapi.azureedge.net/api/M_Est_tob_curr?$filter=Dim1%20eq%20%27SEX_BTSX%27"

import requests as requests
import pandas as pd

#Use request module to get the API data:
response = requests.get(url)

#Turn response into JSON data, using the response function from the requests module:
json_data = response.json()

#Check that the data has been imported and give an error message if not:
if response.status_code == 200:
    print("Data was retrieved")
else:
    print("There was an error retrieving the data")
```

```python
import numpy as np

#Loop through JSON to index, values and years values
index = []
value = []
time = []
for i in json_data["value"]:
    index.append(i["SpatialDim"])
    value.append(i["Value"])
    time.append(i['TimeDimensionValue'])
data = list(zip(value, time))
label = ["Smoking rate", "Year"]
dfqual = pd.DataFrame(data, index = index, columns = label)
dfqual.index.name = "Code"
```

```python
# Import list of countries from csv
import pandas as pd
csv_data = pd.read_csv("/Users/rhonabaker/CFG/GROUP PROJECT/GitHub/GitHub repository/WHO API/WHO_country_list.csv")

country_list = pd.DataFrame(csv_data)
country_list.set_index("Code", inplace = True)
country_list
```

```python
# Merge data into a combined DataFrame
merged_data = pd.merge(dfqual, country_list, left_index=True, right_index=True)
print(merged_data.to_string())
```
