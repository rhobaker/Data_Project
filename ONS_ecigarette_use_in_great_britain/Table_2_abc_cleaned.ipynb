```python
import pandas as pd

#Table 2a
table2a = pd.read_csv(r"C:\Users\Bonald\Desktop\Group project\ONS e-cig use in GB\Table_2a-Table 1.csv", header=1)

table2a.head(15)
```

```python
#Extract and concatenate daily and occasional users

table2a = pd.concat([table2a.loc[0:1], table2a.loc[6:7], table2a.loc[12:13]])

table2a
```

```python
#Combine the rows for daily and occasional user for each year 
 
#First check the datatypes
table2a.dtypes

#The values are numbers, change numeric columns to float values 
numeric_columns = ['Men \r\nCigarette smoker \r\n[note 2]', 'Men \r\nEx-smokers \r\n[note 3]', 'Men \r\nNever smoked \r\n[note 4]', 'Women \r\nCigarette smoker \r\n[note 2]', 'Women \r\nEx-smokers \r\n[note 3]', 'Women \r\nNever smoked \r\n[note 4]', 'All persons \r\nCigarette smoker \r\n[note 2]', 'All persons \r\nEx-smokers \r\n[note 3]', 'All persons \r\nNever smoked \r\n[note 4]']
table2a[numeric_columns] = table2a[numeric_columns].astype(float)

table2a.dtypes
```

```python
#Combine the values in the daily and occasional user columns using the group by, aggregate and sum functions

table2a = table2a.groupby('Year').agg(sum)

table2a
```

```python
#This has made the format of the table strange, with a missing index and second header row. Fix by writing to csv then reading in the table again

table2a.to_csv(r'C:\Users\Bonald\Desktop\Group project\Table2a_cleaned.csv')

table2a_new = pd.read_csv(r"C:\Users\Bonald\Desktop\Group project\Table2a_cleaned.csv")

table2a_new

#The data is now in the expected format. Re-sort to descending year order 
table2a_sorted = table2a_new.sort_values(by = 'Year', ascending=False) 

#Reset the index
table2a_sorted.reset_index()
```

```python
#Table2b

table2b = pd.read_csv(r"C:\Users\Bonald\Desktop\Group project\ONS e-cig use in GB\Table_2b-Table 1.csv", header=1)

table2b.head(15)

#Extra current user rows

table2b = table2b[0:7:6]

table2b
```

```python
#Table 2c
table2c = pd.read_csv(r"C:\Users\Bonald\Desktop\Group project\ONS e-cig use in GB\Table_2c-Table 1.csv")

table2c.head(15)

#Extract current user rows

table2c = table2c.loc[0:18:6]

table2c
```

```python
#Concatenate the tables vertically to combine them
table2 = pd.concat([table2a_sorted, table2b, table2c], axis=0)

#reset the index
table2.reset_index(drop=True, inplace=True)

#View and check full table
table2
```

```python
#Tables haven't combined properly, due to differences in column names

#Rename columns and retry 

columns = ['Year', 'Population proportion e-cig users', 'Men smokers', 'Men ex-smokers', 'Men never smoked', 'Women smokers', 'Women ex-smokers', 'Women never smoked', 
           'All smokers', 'All ex-smokers', 'All never smoked']

table2a_sorted.columns = columns 
table2b.columns = columns
table2c.columns=columns

table2 = pd.concat([table2a_sorted, table2b, table2c], axis=0)

table2.reset_index(drop=True, inplace=True)

table2
```

```python
#Remove the current user column 

table2.drop(['Population proportion e-cig users'], axis=1, inplace=True)

table2.head()
```

```python
#Check for missing values and duplicates

missing = table2.isnull().sum()
print(missing)

duplicates = table2.duplicated().sum()
print(duplicates)
```

```python
#Normalise data types 

#Check data types 
table2.dtypes

# All the columns showing proportion are object data type, we need to change these to float
# Extract the float columns and apply type change 
float_columns = table2.select_dtypes(include='object').columns
table2[float_columns] = table2[float_columns].astype("float")

# Check
table2.dtypes
```

```python
#Explore the proportional numeric data

table2.describe()

#.describe() for the whole table is overwhelming - view summary statistics just for all
table2[["All smokers", "All ex-smokers", "All never smoked"]].describe()

#The distribution for never smoked is skewed by higher values, with more range between quartiles 2 and 3 than quartiles 1 and 2 and the higher values 
#bringing up the mean, but the figures in general are very low so shouldn't have much impact
#All smokers is more evenly spread in quartiles 1-3, but the v high max value is pulling the mean up to near Q3
#Ex-smokers has a generally even distribution
```

```python
#Test visualisations

import matplotlib.pyplot as plt

#E-cig users by smoking status 
year = table2['Year']
smokers = table2['All smokers']
ex = table2['All ex-smokers']
never = table2['All never smoked']

plt.plot(year, smokers, 'g', label='Smokers', linewidth=3)
plt.plot(year, ex, 'r', label='Ex-smokers', linewidth=3)
plt.plot(year, never, 'b', label='Never smoked', linewidth=3)

plt.xlabel("Year")
plt.ylabel("Proportion of e-cig users")
plt.title("Proportion of population as e-cig users by smoking status")

plt.legend()
plt.show()
```

We can see a steep rise in the number of smokers using e-cigs, suggesting that e-cig use isn't necessarily a widely effective tool for quitting. There is however a very steep rise/high skew in the smokers data from 2020 onwards, which may be more extreme due to the survey question change caveats or age category differences as discussed in the 'Table_1abc_cleaned' notebook. We do not have the data on e-cig user smoking status by age, which could show us that younger age groups are skewing the data high, as in the spike in general e-cig use. 

E-cig use of those never smoked has increased, but not to a significant amount - again it would be interesting to see this data by age group. The ex-smokers data shows a relatively steady gradual rise but can be difficult to interpret - the proportion of the population who quit smoking increases each year which will affect the trend, and we cannot tell whether the snapshot of ex-smokers are using e-cigarettes short-term as a part of the quitting process or using them long-term.
