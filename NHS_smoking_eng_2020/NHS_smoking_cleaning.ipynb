#### Data cleaning for NHS datasets - Table S.3.1 regarding the affordability of tobacco in the UK, and Table S.3.2 regarding household expenditure on tobacco

**Table S.3.1**	Indices of tobacco price, retail prices, tobacco price index relative to retail prices index (all items), real disposable income per adult and affordability of tobacco in the United Kingdom


```python
#load the data into a dataframe and view a sample

import pandas as pd

affordability = pd.read_csv("Table S.3.1-Table 1.csv")
affordability
```

**Table S.3.2**	Household expenditure on tobacco (current price value), United Kingdom

```python
expenditure = pd.read_csv("Table S.3.2-Table 1.csv")
expenditure
```

```python
#View information about the dataset

affordability.shape
affordability.info(verbose=True)
#This dataset appear to be clean and in a workable condition
```

```python
expenditure.shape
expenditure.info(verbose=True)
#This dataset appears to have an unnamed column
```

```python
#Check the column names 
sorted(expenditure)

#We can see column 4 is an unnamed blank column - we'll drop this column
expenditure.drop(expenditure.columns[expenditure.columns.str.contains('Unnamed')], axis=1, inplace=True)

#Check that column has been dropped
sorted(expenditure)
```

```python
#I want to directly compare the two datasets, but the expenditure dataset goes back to 1985
# while the affordability dataset only goes to 1987

#For comparison purposes I'll delete the first two rows from the expenditure dataset
expenditure.drop([0, 1], inplace=True)

#Check that rows have been dropped
expenditure.head
```

```python
expenditure.columns

#I want to shorten the column names for simplicity
expenditure.columns = ["Year", "Tobacco Expenditure", "Total Expenditure", "Tobacco Expenditure (%)"]

expenditure.head(5)
```

```python
affordability.columns

#I'll again short the columns for the affordability dataset
affordability.columns = ["Year", "Tobacco Price Index", "Retail Price Index", "TPI relative RPI", "Real Household Disposable Income", "Tobacco Affordability Index"]
affordability.head(6)
```

```python
#Normalising data
expenditure.dtypes

#The tobacco and total expenditure columns in the expenditure dataset are string values, which aren't suitable for analysis
#First I'll remove non numeric values from the columns
expenditure['Tobacco Expenditure'] = expenditure['Tobacco Expenditure'].str.replace(",", "")
expenditure['Total Expenditure'] = expenditure['Total Expenditure'].str.replace(",", "")
expenditure
```

```python

#They I'll convert the data from string to integer
expenditure["Tobacco Expenditure"] = expenditure["Tobacco Expenditure"].astype('int64')
expenditure["Total Expenditure"] = expenditure["Total Expenditure"].astype('int64')
expenditure.dtypes
```

```python
affordability.dtypes
#The affordability data does not need further normalisation
```

```python
#I want to examine the relationship between affordability of tobacco and expenditure on tobacco
#First I'll use the numpy library to find the correlation coefficient
import numpy as np

#Choose the columns I want to compare and convert them into a numby array
afford = affordability["Tobacco Affordability Index"] 
expend = expenditure["Tobacco Expenditure (%)"]
afford_array = np.array(afford)
expend_array = np.array(expend)

#
tobacco_coeff = np.corrcoef(afford_array, expend_array)

#print the correlation coefficient
tobacco_coeff

#Looks like there's a strong positive correlation

```

```python
#I'll see how the data looks on a scatter graph

import matplotlib.pyplot as plt
afford = affordability["Tobacco Affordability Index"] 
expend = expenditure["Tobacco Expenditure (%)"]
plt.scatter(x=afford, y=expend)
plt.xlabel("Tobacco Affordability Index")
plt.ylabel("% of Total Household Expenditure on Tobacco")
plt.title("Relationship between affordability of tobacco and household expenditure on tobacco in the UK (1987-2019)")

#Again we can see the correlation with a few outliers. Maybe we can pinpoint the outliers with a line graph?
```

```python
#I'll use the axes.twinx method to create a line graph that uses two different x-axes while sharing the same y-axis
#This allows me to directly visually compare the two datasets even though they use different units of measurement

#Get the data for the x axis
year = affordability["Year"]

#Plot the first y axis
fig, ax1 = plt.subplots()
color = 'tab:red'
ax1.set_xlabel("Year (1987-2019)")
ax1.set_ylabel("Tobacco Affordability Index", color=color)
ax1.plot(year, afford, color=color, marker='o', markersize=3)
ax1.tick_params(axis='y', labelcolor=color)


#Create a second y axis that shares the same x axis
ax2 = ax1.twinx()  

#Plot the second y axis
color = 'tab:blue'
ax2.set_ylabel("% of Total Household Expenditure on Tobacco", color=color)  
ax2.plot(year, expend, color=color, marker='o', markersize=3)
ax2.tick_params(axis='y', labelcolor=color)

plt.show()
```

It appears that the datasets generally correlates, but there are certain years where household expenditure on tobacco drops despite the affordability remaining relatively high.
Of particular note is the drop in expenditure on tobacco from 2005 to 2008, before it rises again in 2009. 

In 2005 the Government published the Health Bill which sets out proposals for making all premises which are open to the public, including workplaces smokefree. The bill is passed in 2006, when smokefree legislation is implemented in Scotland. In 2007, England, Wales and Northern Ireland all go smokefree.
That same year, the Chancellor announces that the VAT on stop smoking aids will be reduced to the lowest level (5% instead of 17.5%) from 1 July to coincide with the implementation of the indoor smoking ban.

source: https://ash.org.uk/uploads/Key-Dates_2022-04-21-101255_ajre.pdf?v=1650535975

The increase in expenditure on smoking in the years following the smoking ban suggests that while severe legislation such as this can have a noticeable short-term effect, the effect is not long lasting.
